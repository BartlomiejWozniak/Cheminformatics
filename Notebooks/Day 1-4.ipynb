```python
from rdkit import Chem
from rdkit.Chem import rdFingerprintGenerator #Ładuje narzędzie do tworzenia fingerprintów, czyli liczbowych reprezentacji molekuł.
from rdkit.DataStructs import TanimotoSimilarity #Ładujesz funkcję do porównywania fingerprintów.
import pandas as pd #Ładuje pandas, czyli bibliotekę do tabel.
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem.Draw import MolsToGridImage
from rdkit.Chem import Draw
from IPython.display import display
#Importuję podstawowe narzędzia potrzebne mi do kodu
molecules = [       #Tworze listy cząsteczek, para "nazwa", "smiles"
    ("ethanol", "CCO"),

    ("propanol", "CCCO"),

    ("butanol", "CCCCO"),

    ("benzene", "c1ccccc1"),

    ("phenol", "c1ccccc1O")
]
morgan_gen = rdFingerprintGenerator.GetMorganGenerator( #Tworzę generator fingerprintów Morgan
    radius=2, #Radius oznacza promień w którym kod patrzy na otoczenie atomów do 2 kroków
    fpSize=1024 #Fingerprint ma 1024 bitów
)
mols=[] #Pusta lista do ktorej bede dodawac obrazki molekul
fps=[] #Pusta lista dla fingerprintów do których załączam fingerprint wytworzone w petli niżej która określa je dla kazdej czasteczki w "molecules"
names=[] #Pusta lista dla nazw związków dodaje do niej nazwy z "molecules"
similarity_matrix = [] # Pusta lista, zrobiona specjalnie aby powstała tabela podobieństw


for name, smiles in molecules:#Pętla w zbiore "molecules" wyciąga nazwe zwiazku i jego smiles
    mol= Chem.MolFromSmiles(smiles) #Zamiana Smiles kazdej czasteczki w "molecules" na obiekt Molekul
    fp =morgan_gen.GetFingerprint(mol) #Aby wyciągnac fingerprint musze użyc obiektu molekuł stad tutaj korzystam ze zmiennej "mol". Zmiana Molekuły na Wektor Bitów :D
    fps.append(fp)  #do pustej listy dodajy wygenerowane fingerprinty (wektory bitów) z "molecules"
    names.append(name) #Do pustej listy name dodaje nazwe kazdego zwiazku z "molecules" 
    mols.append(mol) # Dodaje do pustej listy wygenerowane obiekty molekul z smiles (do wyswietlenia obrazków)
for fp1 in fps: #Tu bedzie petla w petli, pierwsza petla bierze jeden fingerprint z list fps czyli listy do ktorej dodalem fingerprinty
    row =[] #Tworze pusty wiersz tabeli
    for fp2 in fps: #Petla wewnetrzna która porównuje fp1 ze wszystkimi fingerprinatmi w fps
        similarity = TanimotoSimilarity( #Liczy podobieństwo dwóch finger printów
            fp1,
            fp2)
        row.append(similarity) #Dodaje zmienna "similarity" do utworzonego pustego wiersza tabeli
        
    similarity_matrix.append(row) #Dodanie wiersza do całej tabelii podobieństw musi byc w petli zewnetrznej 

img =MolsToGridImage(mols, #Wytwarzanie obrazow molekul do tabeli
               molsPerRow=5, #Ile molekuł w wierszu
                    subImgSize=(200, 200), #Wielkość
              )
df_similarity = pd.DataFrame( #pandas robi nam tabele nazwana df_similarity
    similarity_matrix,
    index=names, #Nazwa wierszy
    columns=names #Nazwa kolumn
)
display(df_similarity.style.background_gradient(cmap="Reds")) #Wyswietlam tabele z formatem ;)
display(img) #Wyswietlam obrazy molekul
#NAJWAZNIEJSZY SENS SMILES->MOLEKULA->FINGERPRINT->PODOBIENSTWO->TABELA
```
