```python
from rdkit import Chem
from rdkit.Chem import Descriptors
#i lunched the new modul with chemical properties 
molecules =["COCCC", "CCCCCCCC", "c1ccccc1"]
#i made the list for the all compounds like before
for cząsteczki in molecules:
#this is loop to take the next SMILES
    mol =Chem.MolFromSmiles(cząsteczki)
#Changing Text(smiles) into Molecule
    print("SMILES", cząsteczki)
#Printing the chemical formula of the compound ( withouth hydrogen )
    print("Atoms", mol.GetNumAtoms())
    print("MolecularWeight", Descriptors.MolWt(mol))
    print("logP", Descriptors.MolLogP(mol))
#Descriptors was added, from the mol where is Molecule not just text( smiles )
    print("---------------------------")
```

```python
#same like last but i added drawings to the loop
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem import Draw
#i lunched the new modul with chemical properties 
molecules =["COCCC", "CCCCCCCC", "c1ccccc1"]
#i made the list for the all compounds like before
for cząsteczki in molecules:
#this is loop to take the next SMILES
    mol =Chem.MolFromSmiles(cząsteczki)
#Changing Text(smiles) into Molecule
    print("SMILES", cząsteczki)
#Printing the chemical formula of the compound ( withouth hydrogen )
    print("Atoms", mol.GetNumAtoms())
    print("MolecularWeight", Descriptors.MolWt(mol))
    print("logP", Descriptors.MolLogP(mol))
    print("STRUCTURE")
    display(Draw.MolToImage(mol))
#Descriptors was added, from the mol where is Molecule not just text( smiles )
    print("---------------------------")
```
