```python
from rdkit import Chem

#dodałem Chem z biblioteki rdkit ( jest jescze Chem/Draw/Descriptors/DataStructures )
#Chem zawiera podstawowe operacje na molekulach
#Draw

chickennugget =Chem.MolFromSmiles("c1ccccc1")

#mol to tylko nazwa zmiennej moge uzywac sobie jakiekolwiek nazwy 
#po lewej stronie mam zmienna i po prawej Chem.MolFromSmiles oznacza wywołanie funkcji "stwórz molekułe z SMILES" pomiedzy apostrofami jest STRING czyli tekst w jezyku pythona 
#dzieki tej funkcji RdKit czyta "c1ccccc1", interpretując strukturę oraz tworzy obiekt molekuły
#zmienna chickennugget teraz przechowuje atomy,wiązania,strukturę chemiczną,funkcje związane z molekułą.

print(chickennugget.GetNumAtoms())

#tutaj trzeba wywolac obiekt ( w tym przypadku chickenugget ktory odpowiada benzenowi ) GetNumAtoms "Get number of Atoms"- podaje liczbe atomow a print jest potrzebny zeby wyswietlicz wynik
#zasada dzialania jest taka na poczatku liczy nam liczbe atomow a pozniej, wchodzi działanie print i drukuje nam wynik 

print(chickennugget.GetNumBonds())

#tu jest to samo co wyzej tylko że GetNumBonds "Get Number of Bonds"- podaje liczbe wiazan
```
